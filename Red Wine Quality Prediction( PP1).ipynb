```python
x=df.drop('quality',axis=1)
y=df['quality']
```

```python
x
```

```python
y
```

```python
# Balancing the dataset using the SMOTE
import imblearn
from imblearn.over_sampling import SMOTE
x_smote,y_smote = SMOTE().fit_resample(x,y)
y_smote.value_counts()
```


Following the application of SMOTE, the count of both class values has been equalized, indicating a balanced dataset suitable for model creation.

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.decomposition import PCA
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
max_Acc = 0
max_RS = 0

for i in range (1,200):
    x_train, x_test, y_train, y_test = train_test_split(x_smote, y_smote, test_size = 0.20, random_state=i)
    lr = LogisticRegression()
    lr.fit(x_train,y_train)
    predlr = lr.predict(x_test)
    acc = accuracy_score(y_test, predlr)
    if acc>max_Acc:
        max_Acc = acc
        max_RS = i
print("LogisticRegression model")        
print ('Best Accuracy is ', max_Acc, " on Random_State ", max_RS)
```

```python
#Now splitting the dataset into the Training set and test set
x_train,x_test,y_train,y_test=train_test_split(x_smote,y_smote,test_size=0.20,random_state=46)
```

```python
print('Size of X_train : ', x_train.shape)
print('Size of y_train : ', y_train.shape)
print('Size of X_test : ', x_test.shape)
print('Size of Y_test : ', y_test.shape)
```

# Feature Scaling

```python
#feature scaling
st=StandardScaler()
x_train=st.fit_transform(x_train)
x_test =st.transform(x_test)
```

Feature scaling has been performed using the standard scaler to mitigate bias-related issues in the data.

```python
#Using PCA (Principal component analysis)for dimensionality reduction
pca=PCA(n_components=0.90)
x_train=pca.fit_transform(x_train)
x_test =pca.transform(x_test)
```

```python
sum(pca.explained_variance_ratio_)
```

# Logistic Regression

```python
# Now applying Logistic regression
log= LogisticRegression()
log.fit(x_train,y_train)
logPred=log.predict(x_test)
cfr_report = pd.DataFrame(classification_report(y_test,logPred,output_dict=True))

print(f"Classification Report\n{cfr_report}\n")
print(f" Confusion Matrix:\n{confusion_matrix(y_test,logPred)}\n")
print("Accuracy score",accuracy_score(y_test,logPred))
score = cross_val_score(log,x,y,cv=5)
print("Cross val score: ", score.mean())
```

# Support Vector Classifier

```python
svm = SVC()
svm.fit(x_train,y_train)
svmPred = svm.predict(x_test)
cfr_report = pd.DataFrame(classification_report(y_test,svmPred,output_dict=True))

print(f"Classification Report\n{cfr_report}\n")
print(f" Confusion Matrix:\n{confusion_matrix(y_test,svmPred)}\n")
print("Accuracy score",accuracy_score(y_test,svmPred))
score1 = cross_val_score(svm,x,y)
print("Cross val score:",score1.mean())
```

# DecisionTreeClassifier

```python
dt =DecisionTreeClassifier()
dt.fit(x_train,y_train)
dtPred= dt.predict(x_test)
cfr_report = pd.DataFrame(classification_report(y_test,dtPred,output_dict=True))

print(f"Classification Report\n{cfr_report}\n")
print(f" Confusion Matrix:\n{confusion_matrix(y_test,dtPred)}\n")
print("Accuracy score",accuracy_score(y_test,dtPred))
score2 = cross_val_score(dt,x,y)
print("Cross val score: ", score2.mean())
```


# RandomForestClassifier

```python
rf=RandomForestClassifier()
rf.fit(x_train,y_train)
rfPred = rf.predict(x_test)
cfr_report = pd.DataFrame(classification_report(y_test,rfPred,output_dict=True))

print(f"Classification Report\n{cfr_report}\n")
print(f" Confusion Matrix:\n{confusion_matrix(y_test,rfPred)}\n")
print("Accuracy score",accuracy_score(y_test,rfPred))
score3 = cross_val_score(rf,x,y)
print("Cross val score: ", score3.mean())
```

# GradientBoostingClassifier

```python
gbc= GradientBoostingClassifier()
gbc.fit(x_train,y_train)
gbcPred=gbc.predict(x_test)
cfr_report = pd.DataFrame(classification_report(y_test,gbcPred,output_dict=True))

print(f"Classification Report\n{cfr_report}\n")
print(f" Confusion Matrix:\n{confusion_matrix(y_test,gbcPred)}\n")
print("Accuracy score:",accuracy_score(y_test,gbcPred))
score4 = cross_val_score(gbc,x,y)
print("Cross val score: ", score4.mean())
```

# KNeighborsClassifier

```python
knn =KNeighborsClassifier()
knn.fit(x_train,y_train)
knnPred = knn.predict(x_test)
cfr_report = pd.DataFrame(classification_report(y_test,knnPred,output_dict=True))

print(f"Classification Report\n{cfr_report}\n")
print(f" Confusion Matrix:\n{confusion_matrix(y_test,knnPred)}\n")
print("Accuracy score",accuracy_score(y_test,knnPred))
score5 = cross_val_score(knn,x,y)
print("Cross val score: ", score5.mean())
```

```python
from sklearn.metrics import roc_curve,roc_auc_score
from sklearn.metrics import roc_curve
```

```python
# Compute ROC curves for each model
fpr_dt, tpr_dt, _ = roc_curve(y_train, dt.predict_proba(x_train)[:, 1])
fpr_log, tpr_log, _ = roc_curve(y_train, log.predict_proba(x_train)[:, 1])
fpr_svm, tpr_svm, _ = roc_curve(y_train, svm.decision_function(x_train))
fpr_rf, tpr_rf, _ = roc_curve(y_train, rf.predict_proba(x_train)[:, 1])
fpr_knn, tpr_knn, _ = roc_curve(y_train, knn.predict_proba(x_train)[:, 1])
fpr_gbc, tpr_gbc, _ = roc_curve(y_train, gbc.predict_proba(x_train)[:, 1])

# Plot ROC curves
plt.figure(figsize=(6, 4))
plt.plot(fpr_dt, tpr_dt, label='Decision Tree')
plt.plot(fpr_log, tpr_log, label='Logistic Regression')
plt.plot(fpr_svm, tpr_svm, label='SVM')
plt.plot(fpr_rf, tpr_rf, label='Random Forest')
plt.plot(fpr_knn, tpr_knn, label='KNN')
plt.plot(fpr_gbc, tpr_gbc, label='Gradient Boosting')

# Add labels and legend
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curves for Different Models')
plt.legend(prop={'size': 10}, loc='lower right')

# Show plot
plt.show()
```

# Hyperparameter Tuning

```python
from sklearn.model_selection import GridSearchCV
 
# Creating the hyperparameter grid
param_grid = {'solver': ['newton-cg', 'lbfgs', 'liblinear','saga','sag'],
              'penalty': ['l1','l2'],
              'C':[0.0001, 0.001, 0.01, 0.1, 1, 10, 100, 1000],
             'dual':[True,False],
            'max_iter': [100,110,120,130,140]}

# Instantiating logistic regression classifier
model = LogisticRegression()
# Instantiating the GridSearchCV object
gsv = GridSearchCV( estimator = model,  param_grid = param_grid, cv = 5)
gsv.fit(x_train, y_train)
 
# Print the tuned parameters and score
print("Tuned Logistic Regression Parameters: {}".format(gsv.best_params_))
print("Best score is {}".format(gsv.best_score_))
```
